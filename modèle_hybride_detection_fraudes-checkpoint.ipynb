```python
#partie 1 modèle non supervisé_SOM

#librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#importer les données
dataset=pd.read_csv("Downloads/Credit_Card_Applications.csv")
#Segmentation des clients
#un neurone par variable
X=dataset.iloc[:,:-1].values #toutes les colonnes sauf la dernière
y=dataset.iloc[:,-1].values

#si on manipule des distances entre neurones ont doit changer l'echelle des variables
from sklearn.preprocessing import MinMaxScaler #transformer les variables [0-1]
sc=MinMaxScaler(feature_range = (0, 1))
X=sc.fit_transform(X)

#créer la carte SOM
from minisom import MiniSom

#x,y dimension de la carte x*y=100
#input_len=nb de colonnes du jeu de données
som=MiniSom(x=10,y=10, input_len=15)

#initialiser les poids 
som.random_weights_init(X)

#entrainer notre SOM
som.train_random(X,num_iteration=100)

#visualisation de la carte 
from pylab import bone,pcolor,colorbar,plot,show
bone()
pcolor(som.distance_map().T)
colorbar()

#juxtaposer les données des clients acceptés et refusés du jeu de données Y sur la carte 
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']# vecteur de marqueurs
colors = ['r', 'g']# r=refusé, g=accepté
for i, x in enumerate(X):
    w = som.winner(x) #récupérer le neurone gagnant de la ligne (observation) du client
    plot(w[0] + 0.5, #w[0] abscisse, +0.5 --> au centre
         w[1] + 0.5,
         markers[y[i]], #markers[y[i]]= marker en rond ou en carré les classes de sorties (est ce que les clients sont accep^tés par la banque ou pas)
         markeredgecolor = colors[y[i]], #colorier la bordure des cercles et rond par rouge en cas de refus ou vert en cas d'acceptation
         markerfacecolor = 'None', #ne pas colorier l'interieur 
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mapping=som.win_map(X)
fraudes=np.concatenate((mapping[(1,7)],mapping[(2,7)],mapping[(7,7)]),axis=0)
#récupérer la liste des clients frauduleux
fraudes=sc.inverse_transform(fraudes)
fraudes
```

```python
#partie 2 passer du non supervisé au supervisé
clients=dataset.iloc[:,1:].values #on ne prend pas l'id custmer
is_fraud=np.zeros(len(dataset))       
```

```python
for i in range(len(dataset)):
    if dataset.iloc[i,0] in fraudes: #i: indice de ligne, 0 : colonne id_custmer
        is_fraud[i]=1
```

```python
 # Encoder les données
from sklearn.preprocessing import  StandardScaler

sc=StandardScaler()
clients= sc.fit_transform(clients)

# Importer les librairies Keras
from keras.models import Sequential
from keras.layers import Dense 

 # Initialiser notre ANN
classifier = Sequential()

# Ajouter les couches d'entrées et cachée
classifier.add(Dense(units = 2, kernel_initializer = 'uniform', activation = 'relu', input_dim = 15))

# Ajouter une couche de sortie
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiler notre ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# entrainement de notre ANN
classifier.fit(clients, is_fraud, batch_size = 1, epochs = 2)

 # Prediction 
y_pred = classifier.predict(clients)
#y_pred = (y_pred > 0.5)
```
