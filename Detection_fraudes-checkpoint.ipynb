```python
#librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importer les données
dataset=pd.read_csv("Downloads/Credit_Card_Applications.csv")
#Segmentation des clients
#un neurone par variable
X=dataset.iloc[:,:-1].values #toutes les colonnes sauf la dernière
y=dataset.iloc[:,-1].values
```

```python
#si on manipule des distances entre neurones ont doit changer l'echelle des variables
from sklearn.preprocessing import MinMaxScaler #transformer les variables [0-1]
sc=MinMaxScaler(feature_range = (0, 1))
X=sc.fit_transform(X)
```

```python
#créer la carte SOM
from minisom import MiniSom

#x,y dimension de la carte x*y=100
#input_len=nb de colonnes du jeu de données
som=MiniSom(x=10,y=10, input_len=15)

#initialiser les poids 
som.random_weights_init(X)

#entrainer notre SOM
som.train_random(X,num_iteration=100)
```

```python
#visualisation de la carte 
from pylab import bone,pcolor,colorbar,plot,show
bone()
pcolor(som.distance_map().T)
colorbar()
```

```python
#juxtaposer les données des clients acceptés et refusés du jeu de données Y sur la carte 
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']# vecteur de marqueurs
colors = ['r', 'g']# r=refusé, g=accepté
for i, x in enumerate(X):
    w = som.winner(x) #récupérer le neurone gagnant de la ligne (observation) du client
    plot(w[0] + 0.5, #w[0] abscisse, +0.5 --> au centre
         w[1] + 0.5,
         markers[y[i]], #markers[y[i]]= marker en rond ou en carré les classes de sorties (est ce que les clients sont accep^tés par la banque ou pas)
         markeredgecolor = colors[y[i]], #colorier la bordure des cercles et rond par rouge en cas de refus ou vert en cas d'acceptation
         markerfacecolor = 'None', #ne pas colorier l'interieur 
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mapping=som.win_map(X)
fraudes=np.concatenate((mapping[(4,2)],mapping[(4,5)],mapping[(5,3)]),axis=0)
```

```python
#récupérer la liste des clients frauduleux
fraudes=sc.inverse_transform(fraudes)
fraudes
```
